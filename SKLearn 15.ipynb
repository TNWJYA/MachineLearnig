### Terminology: Root Node, Internal Node, Leaf Node

- Contoh Decision Tree (Pohon Keputusan)

- Decision Tree (Pohon Keputusan) adalah metode clasisfikasi yang menerapkan struktur pohon

- Root node

    * Root node atau biasa disebut dengan *Root(akar)*, karena pada contoh struktur decision tree diatas terbalik maka           root node menempati posisi paling atas. 
    * pertanyaan __Fear technology?__ merupakan root node
    
- leaf node

    * Leaf node merupakan sejumlah node yang posisi nya diujung bawah, dikarenakan contoh decision tree diatas terbalik           maka komponen ini disebut leaf node.
    * pada contoh decision tree diatas terdapat 5 buah leaf node yaitu __Apple, Chrome, Ubuntu linux, Arch linux dan             Windows__
    * Setiap leaf dari decision tree  akan merepresentasikan prediksi class yang akan dihasilkan oleh struktur tree               tersebut.
    
    
- Internal node
  
  * internal node merupakan sejumlah node yang berada di antara leaf dan root
  * pada contoh diatas terdapat 3 buah internal node yaitu pertanyaan __Your dad rich?__, __Care about privacy?__ dan         __Have life?__

### Membangun Decision Tree

## Mempersiapkan Dataset

```python
from sklearn.datasets import load_iris

X, y = load_iris(return_X_y=True)

print(f'Dimensi Feature: {X.shape}')
print(f'Class: {set(y)}')
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X,
                                                    y,
                                                    test_size=0.3,
                                                    random_state=0)
```

### Classification dengan DecisionTreeClassifier

```python
from sklearn.tree import DecisionTreeClassifier

model = DecisionTreeClassifier(max_depth=4)

model.fit(X_train, y_train)
```

### Visualisasi Model

```python
import matplotlib.pyplot as plt
from sklearn import tree

plt.rcParams['figure.dpi'] = 85
plt.subplots(figsize=(10, 10))
tree.plot_tree(model, fontsize=10)
plt.show()
```

### Evaluasi Model

```python
from sklearn.metrics import classification_report

y_pred = model.predict(X_test)

print(classification_report(y_test, y_pred))
```
